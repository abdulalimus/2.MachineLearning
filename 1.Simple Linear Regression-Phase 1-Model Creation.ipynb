```python
# Pandas is a library that contains the information related to database
import pandas as pd
```

```python
# Importing the dataset from a CSV file into a DataFrame
dataset = pd.read_csv("Salary_Data.csv")
dataset
```

```python
# Extracting the 'YearsExperience' column as the independent variable (input) in a DataFrame format
independent = dataset[["YearsExperience"]]
independent
```

```python
# Extracting the 'Salary' column as the dependent variable (target or output) in a DataFrame format
dependent = dataset[["Salary"]]
dependent
```

```python
# Importing the train_test_split function from sklearn to split the dataset
from sklearn.model_selection import train_test_split
# Splitting the dataset into training and testing sets (70% training, 30% testing)
X_train, X_test, Y_train, Y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#learning phase
# Importing the LinearRegression model
from sklearn.linear_model import LinearRegression
# Creating an instance of the LinearRegression model
regressor = LinearRegression()
# Fitting (training) the model using the training data
regressor.fit(X_train,Y_train)
```

```python
# Extracting the coefficient (weight) of the linear regression model
weight = regressor.coef_
weight
```

```python
# Extracting the intercept (bias) of the linear regression model
bias = regressor.intercept_
bias
```

```python
#Making predictions on the test set using the trained linear regression model
Y_predict = regressor.predict(X_test)
```

```python
# Importing the r2_score function to evaluate model performance
from sklearn.metrics import r2_score
# Calculating the R-squared score to assess the accuracy of the predictions - score near 1 means AI is perfroming well
r_score = r2_score(Y_test, Y_predict)
r_score
```

```python
# Importing the pickle library to save and load Python objects
import pickle

# Defining the filename where the trained model will be saved
filename = "finalized_model_linear.sav"
```

```python
# Saving the trained linear regression model to a file using pickle
pickle.dump(regressor, open(filename,'wb'))
```

```python
# Loading the previously saved linear regression model from the file using pickle
loaded_model = pickle.load(open("finalized_model_linear.sav", 'rb'))

# Using the loaded model to make a prediction for a new input (e.g., 11 years of experience)
result = loaded_model.predict([[11]])
result
```
