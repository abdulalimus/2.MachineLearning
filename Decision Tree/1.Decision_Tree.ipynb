```python
# Pandas is a library that contains the information related to database
import pandas as pd
```

```python
# Importing the dataset from a CSV file into a DataFrame
dataset = pd.read_csv("50_Startups.csv")
dataset
```

```python
# One Hot encoding - Column expansion; drop the first expanded column i.e. State_California to improve the API performance
dataset = pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
# Extracting the columns as the independent variable (input) in a DataFrame format
independent = dataset[['R&D Spend', 'Administration', 'Marketing Spend',
       'State_Florida', 'State_New York']]
independent
```

```python
# Extracting the 'Profit' column as the dependent variable (target or output) in a DataFrame format
dependent = dataset[['Profit']]
dependent
```

```python
# Importing the train_test_split function from sklearn to split the dataset
from sklearn.model_selection import train_test_split
# Splitting the dataset into training and testing sets (70% training, 30% testing)
X_train, X_test, Y_train, Y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#learning phase
# Importing the Decision Tree model
from sklearn.tree import DecisionTreeRegressor
#first iteration - no parameters
#regressor = DecisionTreeRegressor ()
## second iteration - pass the parameters and check if there is improvement
regressor = DecisionTreeRegressor (criterion='absolute_error',splitter='best')

# Fitting (training) the model using the training data; TRAINED MODEL WILL BE SAVED IN regressor variable; 'regressor' is a custom name
regressor = regressor.fit(X_train,Y_train)
```

```python
# to create the graph
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree (regressor)
plt.show ()
```

```python
#Making predictions on the test set using the trained model
Y_predict = regressor.predict(X_test)
```

```python
# Importing the r2_score function to evaluate model performance
from sklearn.metrics import r2_score
# Calculating the R-squared score to assess the accuracy of the predictions - score near 1 means AI is perfroming well
r_score = r2_score(Y_test, Y_predict)
r_score
```

```python
# Importing the pickle library to save and load Python objects
import pickle

# Defining the filename where the trained model will be saved
filename = "finalized_model_DT.sav"
```

```python
# Saving the trained linear regression model to a file using pickle
pickle.dump(regressor, open(filename,'wb'))
```

```python
# Loading the previously saved linear regression model from the file using pickle
loaded_model = pickle.load(open("finalized_model_DT.sav", 'rb'))

# Using the loaded model to make a prediction for a new input
result = loaded_model.predict([[12345.20,23456.80,3456.10,0,0]])
result
```
